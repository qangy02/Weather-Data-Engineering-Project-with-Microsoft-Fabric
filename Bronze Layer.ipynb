#### fetch data from API and save it to Files

```python
# Define the Open Meteo API endpoint and parameters
url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&hourly=temperature_2m"

#Make API request
response = requests.get(url)

#Check if request was successful
if response.status_code == 200:
    weather_data = response.json()

    file_path = f"/lakehouse/default/Files/{city}_weather.json"

    with open(file_path, 'w') as json_file:
        json.dump(weather_data, json_file, indent=4)

    print("Sucessfully feteched Data from Open Meto API!")
else:
    raise Exception(f"Failed to fetch Data from Open Meteo API: {response.status_code} ")
```
